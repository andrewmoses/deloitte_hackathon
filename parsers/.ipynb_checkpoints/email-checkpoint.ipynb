```python
from nltk.corpus import stopwords
sw_list = stopwords.words('english')
import string
remove_punct_dict = dict((ord(punct), None) for punct in string.punctuation)
def remove_stopwords(string,sw_list):
    new_s=[]
    for word in string.split():
        if word not in sw_list:
            new_s.append(word)
    new_s = " ".join(new_s)
    return new_s
from nltk.stem import PorterStemmer
ps = PorterStemmer()
def stemming(string):
    new_s=[]
    for word in string.split():
        s_word = ps.stem(word)
        if len(s_word) > 2:
            new_s.append(s_word)
    return " ".join(new_s)
```

```python
sms_list = []
for x in range(500):
    msg = df['message'][x]
    parsed = msg.split('\n')
    stringed = ""
    for ln in parsed[15:]:
        stringed = stringed.strip() + ' ' + ln.strip()
#     print(stringed)
    p_string = stringed.translate(remove_punct_dict)
    f_string = remove_stopwords(p_string,sw_list)
    f_string = stemming(f_string)
    print(f_string)
    sms_list.append(f_string)
    print("*"*50)
```

```python
from nltk import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
TfidfVec = TfidfVectorizer(tokenizer=word_tokenize)
tfidf_transformer = TfidfVec.fit(sms_list)
tfidf = tfidf_transformer.transform(sms_list)
```

```python
new_df = pd.DataFrame(tfidf.todense())
new_df
```

```python
%matplotlib inline
from sklearn.cluster import KMeans 
from sklearn import metrics 
from scipy.spatial.distance import cdist 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
X = new_df.values
distortions = [] 
inertias = [] 
mapping1 = {} 
mapping2 = {} 
K = range(1,20) 
  
for k in K:
    print("wowo da")
    #Building and fitting the model 
    kmeanModel = KMeans(n_clusters=k).fit(X) 
    kmeanModel.fit(X)     
      
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 
                      'euclidean'),axis=1)) / X.shape[0]) 
    inertias.append(kmeanModel.inertia_) 
  
    mapping1[k] = sum(np.min(cdist(X, kmeanModel.cluster_centers_, 
                 'euclidean'),axis=1)) / X.shape[0] 
    mapping2[k] = kmeanModel.inertia_ 
```

```python
for key,val in mapping1.items(): 
    print(str(key)+' : '+str(val))
```

```python
plt.plot(K, distortions, 'bx-') 
plt.xlabel('Values of K') 
plt.ylabel('Distortion') 
plt.title('The Elbow Method using Distortion') 
plt.show()
```
